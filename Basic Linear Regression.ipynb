````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
````

````python
# Forecasted temperatures (Features)
X = np.array([60, 62, 67, 70, 71, 75, 78, 80, 82, 85]).reshape(-1, 1)

# Actual temperatures (Target) - notice they are slightly different!
y = np.array([62, 63, 66, 72, 70, 77, 77, 81, 84, 87])

print("Data created!")
````

````python
y
````

````python
plt.scatter(X, y, color='blue')
plt.xlabel('Forecasted Temperature (°F)')
plt.ylabel('Actual Temperature (°F)')
plt.title('Forecast vs. Actual')
plt.show()
````

````python
# 1. Create the model
model = LinearRegression()

# 2. Train the model (mapping X to y)
model.fit(X, y)

print("The model has finished learning!")
````

````python
# Use the model to predict values for our line
y_pred = model.predict(X)

# Plot the original dots
plt.scatter(X, y, color='blue', label='Actual Data')

# Plot the "Best Fit" line
plt.plot(X, y_pred, color='red', label='Prediction Line')

plt.xlabel('Forecasted Temp')
plt.ylabel('Actual Temp')
plt.legend()
plt.show()
````

````python
new_forecast = np.array([[83]])
corrected_temp = model.predict(new_forecast)

print(f"Original Forecast: 83°F")
print(f"Model's Corrected Prediction: {corrected_temp[0]:.2f}°F")
````

````python
from sklearn.metrics import mean_squared_error, r2_score

# Check the accuracy
score = r2_score(y, y_pred)
error = mean_squared_error(y, y_pred)

print(f"R-squared Score: {score:.4f}")
print(f"Average 'Miss' (MSE): {error:.4f}")
print(f'"{score:.2}% of the variation in actual temperatures is explained by the model (using forecasted temperatures)."')
````

"98% of the variation in actual temperatures is explained by the model (using forecasted temperatures)."

= "Only 12% of the fluctuations in the actual temperature remain as unexplained residual noise."

````python
import math
rmse = math.sqrt(error)
print(f"Typical Error (RMSE): {rmse:.2f} °F")
print(f'"The typical error between the observed and predicted temperatures is {rmse:.2f} °F."')
````

### 📖 Key Terms & Definitions
- **Variation (Spread):** This refers to the differences or "wiggle" in the **Actual Temperature (y)**. It describes how much the temperatures vary from a flat, unchanging line.
  - **The Mystery:** The actual temperature (y) has a lot of "mystery wiggle." We don't know why it goes up and down.
  - **The Explanation:** We check the **Forecast (X)** to see if its wiggle matches the truth. Since they match so well, we say the Forecast (X) "explains" the mystery.

### 📝 Cast of Characters
1. **X (Forecasted Temp):** This is our **Predictor** (the input). "The forecasted temperature (X) is used to explain the movement of the actual temperature."
2. **y (Actual Temp):** This is our **Observed Truth** (the target). "We are studying the variation in the actual observed temperature (y)."
3. **y_pred (Model Guess):** This is our **Predicted Value** (the output). "The predicted temperatures (y_pred) represent the red line—the best guess the model can make."
4. **R-squared (0.9757):** This is our **Score**. "97.57% of the variation in the actual temperature (y) is explained by the forecasted temp (X)."
5. **MSE / RMSE:** This is our **Error**. "The typical error (typical distance) between the observed (y) and predicted (y_pred) temperatures is 1.29 °F."
